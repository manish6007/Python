# Forthcoming Result Calendar and Corporate Action Details from NSE

Find out dates of forthcoming corporate events like Financial results/Dividend for all the stocks listed on NSE. 
Does not check for internet connection, overwrites CSV file and writes a summary to csv file named boardMeetDates.csv and CorporateActions.csv

# Result Calendar

```python
#Importing Libraries
import requests
import pandas as pd

# Download the CSV file and save it in same folder as this Notebook
url = 'https://www1.nseindia.com/corporates/datafiles/BM_All_Forthcoming.csv'
bmDates = requests.get(url, allow_redirects = True)
open('BM_All_Forthcoming.csv', 'wb').write(bmDates.content)

# Import the CSV file and seach for BM dates for given list of scrips
# Reading CSV file in Pandas
bmDate_CSV = pd.read_csv('BM_All_Forthcoming.csv', index_col=None, encoding='utf-8')
dFA = pd.DataFrame(bmDate_CSV)

#Delete duplicate row entries for same scrips
dFB = dFA.drop_duplicates(subset=None, keep='first', inplace=False)

#Assign a column with unique values as the index (X-Axis) of the dataframe 
dFB = pd.DataFrame(dFB.sort_values(by = ['Symbol']),
                    columns = ['Symbol', 'BoardMeetingDate'])

# Input data file is available in the same dirctory where the notebook is placed.
# Reading the Input file and convert it into the list
scripFile = open("stocksList.txt")
scripArray = scripFile.read()
scripList = scripArray.split("\n")

#Creating the final Dataframe for the script for which the Result date is available on NSE.
dFF = dFB.loc[dFB['Symbol'].isin(scripList)]
scripFile.close

# Storing the data in boardMeetDates.csv
dFF.to_csv("boardMeetDates.csv")
dFF


```

# Corporate Action

```python
#Importing Libraries
import requests
import pandas as pd

# Download the CSV file and save it in same folder as this Notebook
url = 'https://www1.nseindia.com/corporates/datafiles/CA_ALL_FORTHCOMING.csv'
caDates = requests.get(url, allow_redirects = True)
open('CA_ALL_FORTHCOMING.csv', 'wb').write(caDates.content)

# Import the CSV file and seach for BM dates for given list of scrips
# Reading CSV file in Pandas Dataframe
caDates_CSV = pd.read_csv('CA_ALL_FORTHCOMING.csv', index_col=None, encoding='utf-8')
dFA = pd.DataFrame(caDates_CSV)

#Delete duplicate row entries for same scrips
dFB = dFA.drop_duplicates(subset=None, keep='first', inplace=False)

#Assign a column with unique values as the index (X-Axis) of the dataframe 
dFB = pd.DataFrame(dFB.sort_values(by = ['Symbol']),
                    columns = ['Symbol', 'Series','Face Value(Rs.)','Purpose','Ex-Date', 'Record Date', 'BC Start Date', 'BC End Date'])

#Filtering  he EQ series from Data
dFB = dFB[dFB['Series']=='EQ']

# Input data file is available in the same dirctory where the notebook is placed.
# Reading the Input file and convert it into the list
scripFile = open("stocksList.txt")
scripArray = scripFile.read()
scripList = scripArray.split("\n")

#Creating the final Dataframe for the script for which the Result date is available on NSE.
dFF = dFB.loc[dFB['Symbol'].isin(scripList)]
scripFile.close

# Storing the data in boardMeetDates.csv
dFF.to_csv("CorporateActions.csv")
dFF
```
